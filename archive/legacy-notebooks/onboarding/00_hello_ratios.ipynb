## 1. A Pizza is a Portfolio

Imagine a pizza cut into 8 slices, shared among 4 friends.

Each person gets a **share** (ρ) of the whole pizza.  
The **unity constraint** says: all shares must add up to 1 (the whole pizza).  

This is not an assumption — it's a fact. You can't eat more than 100% of one pizza.

```python
# Our pizza portfolio: 4 friends, 8 slices
friends = ['Alice', 'Bob', 'Carol', 'Dave']
slices  = [2, 2, 2, 2]  # Everyone gets 2 slices

# Convert slices to shares (ratios of the whole)
total = sum(slices)
shares = [s / total for s in slices]

print('Slices:', dict(zip(friends, slices)))
print('Shares:', dict(zip(friends, [f'{s:.2f}' for s in shares])))
print(f'Sum of shares: {sum(shares):.2f}  ← Unity constraint satisfied!')
```

```python
# Visualize the equal pizza
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Pie chart
ax1.pie(shares, labels=friends, colors=colors, autopct='%1.0f%%',
        startangle=90, textprops={'fontsize': 14})
ax1.set_title('Pizza Portfolio: Equal Shares', fontsize=16, fontweight='bold')

# Bar chart of shares
ax2.bar(friends, shares, color=colors, edgecolor='white', linewidth=2)
ax2.axhline(y=0.25, color='gray', linestyle='--', alpha=0.5, label='Equal share (0.25)')
ax2.set_ylabel('Share (ρ)', fontsize=14)
ax2.set_title('Each Person\'s Share', fontsize=16, fontweight='bold')
ax2.set_ylim(0, 0.5)
ax2.legend()

plt.tight_layout()
plt.show()
print(f'∑ρ = {sum(shares):.2f}  ✔ Unity constraint holds.')
```

## 2. Someone Takes an Extra Slice

Now Bob sneaks an extra slice. Nobody announced this. Nobody agreed to it.  
This is **silent drift** — the shares changed, but no one declared a reason.

Notice: the unity constraint STILL holds (it always does).  
But the shares are now *different from what was declared*.

```python
# The declared plan: everyone gets equal shares
declared = [0.25, 0.25, 0.25, 0.25]

# What actually happened: Bob took Carol's slice
observed = [2/8, 3/8, 1/8, 2/8]  # Bob got 3, Carol got 1

print('DECLARED shares:', dict(zip(friends, [f'{d:.3f}' for d in declared])))
print('OBSERVED shares:', dict(zip(friends, [f'{o:.3f}' for o in observed])))
print(f'\nSum of observed: {sum(observed):.2f}  ← Unity constraint STILL holds!')
print('\nBut the distribution changed silently...')
```

```python
# Compute the DRIFT GAP for each friend
drift_gaps = [abs(d - o) for d, o in zip(declared, observed)]

# Mean Drift Gap (MDG) — the average drift across all elements
mdg = np.mean(drift_gaps)

print('Drift gaps:')
for f, dg in zip(friends, drift_gaps):
    flag = ' ⚠️ DRIFT!' if dg > 0 else ' ✔'
    print(f'  {f}: |{declared[friends.index(f)]:.3f} - {observed[friends.index(f)]:.3f}| = {dg:.3f}{flag}')

print(f'\n★ Mean Drift Gap (MDG) = {mdg:.3f} = {mdg*100:.1f} percentage points')
print(f'\nInterpretation: MDG > 0 means someone\'s share shifted without a declaration.')
print(f'HUF detected silent drift — Bob\'s gain came at Carol\'s expense.')
```

```python
# Visualize: Declared vs Observed
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

x = np.arange(len(friends))
width = 0.35

# Side-by-side bars
bars1 = ax1.bar(x - width/2, declared, width, label='Declared', color='#45B7D1', edgecolor='white')
bars2 = ax1.bar(x + width/2, observed, width, label='Observed', color='#FF6B6B', edgecolor='white')
ax1.set_xlabel('Friend')
ax1.set_ylabel('Share (ρ)')
ax1.set_title('Declared vs Observed Shares', fontsize=14, fontweight='bold')
ax1.set_xticks(x)
ax1.set_xticklabels(friends)
ax1.legend()
ax1.set_ylim(0, 0.5)

# Drift gap bars
drift_colors = ['#2ECC71' if dg == 0 else '#E74C3C' for dg in drift_gaps]
ax2.bar(friends, [dg * 100 for dg in drift_gaps], color=drift_colors, edgecolor='white')
ax2.axhline(y=mdg*100, color='orange', linestyle='--', linewidth=2, label=f'MDG = {mdg*100:.1f}pp')
ax2.set_ylabel('Drift Gap (percentage points)')
ax2.set_title('Drift Gap per Person', fontsize=14, fontweight='bold')
ax2.legend(fontsize=12)

plt.tight_layout()
plt.show()
print('Red bars = silent drift detected. Orange line = Mean Drift Gap.')
```

## 3. Try It Yourself!

Change the `observed` list below and re-run the cell.  
The 8 slices must still add up to 8 (unity constraint).  

**Questions to explore**:
- What happens if one person gets ALL 8 slices? (Concentration Trap — FM-4)
- What happens if everyone gets exactly 2? (Ground State — MDG = 0)
- What if Alice gets 3 and Bob gets 1? Does MDG stay the same?

```python
# ✏️ CHANGE THESE VALUES and re-run! (Must add up to 8 slices)
my_slices = [2, 3, 1, 2]  # ← Try: [4, 2, 1, 1] or [8, 0, 0, 0] or [2, 2, 2, 2]

# ------- computation (don't change below) -------
assert sum(my_slices) == 8, f'Slices must add to 8! You have {sum(my_slices)}.'
my_shares = [s / 8 for s in my_slices]
my_declared = [0.25, 0.25, 0.25, 0.25]
my_gaps = [abs(d - o) for d, o in zip(my_declared, my_shares)]
my_mdg = np.mean(my_gaps)

print(f'Shares: {dict(zip(friends, [f"{s:.3f}" for s in my_shares]))}')
print(f'\u2211\u03c1 = {sum(my_shares):.2f}  \u2714 Unity constraint holds.')
print(f'MDG = {my_mdg*100:.1f} percentage points')

if my_mdg == 0:
    print('\n\u2728 GROUND STATE! No drift. Perfect alignment.')
elif my_mdg * 100 < 5:
    print(f'\n\u26a0\ufe0f Action window open. Drift detected but small.')
else:
    print(f'\n\ud83d\udea8 Active drift! MDG = {my_mdg*100:.1f}pp > 5pp threshold.')

# Check for concentration
sorted_shares = sorted(my_shares, reverse=True)
cumsum = 0
proof_line = 0
for s in sorted_shares:
    cumsum += s
    proof_line += 1
    if cumsum >= 0.8:
        break
print(f'PROOF line = {proof_line} (elements needed for 80% of portfolio)')
if proof_line == 1:
    print('\ud83d\udea8 Concentration Trap (FM-4)! One person holds 80%+ of the pizza.')
```
